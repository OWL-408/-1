```python
#PDFを結合

from PyPDF2 import PdfMerger
from google.colab import files

# ユーザーにファイルをアップロードさせる
uploaded = files.upload()

# アップロードされたファイルの名前をリスト化
pdf_files = list(uploaded.keys())

# 必要に応じて順序を調整（例えば、リストを逆順にしたい場合）
pdf_files.sort()  # 名前順に並べ替えたい場合（必要に応じて削除可能）
# pdf_files.reverse()  # リバースする場合はこれを使う
# pdf_files = ['file1.pdf', 'file2.pdf', 'file3.pdf']  # 任意の順序で指定


# Mergerを使って結合
merger = PdfMerger()
for pdf in pdf_files:
    merger.append(pdf)

# 結合したPDFを保存
output_filename = "内見希望物件リスト"
merger.write(output_filename)
merger.close()

# 保存したPDFをダウンロード
files.download(output_filename)
```

```python
#指定ページ削除

from PyPDF2 import PdfReader, PdfWriter
from google.colab import files

# ファイルをアップロード
uploaded = files.upload()

# アップロードされたファイルの名前をリスト化
pdf_files = list(uploaded.keys())

# 削除するページ番号（例：2ページ目を削除したい場合は1を指定）
pages_to_remove = [1, 2, 3, 4, 5, 6, 7]  # 0始まりのインデックス (例: [0, 1, 2] なら1ページ目から3ページ目が削除対象)

# 新しいPDFを作成
writer = PdfWriter()

for pdf_file in pdf_files:
    reader = PdfReader(pdf_file)
    num_pages = len(reader.pages)

    # ページを1つずつチェックし、削除対象以外を追加
    for i in range(num_pages):
        if i not in pages_to_remove:
            writer.add_page(reader.pages[i])

# 保存するファイル名
output_filename = "output_without_pages.pdf"
with open(output_filename, "wb") as output_file:
    writer.write(output_file)

# 保存したPDFをダウンロード
files.download(output_filename)
```
